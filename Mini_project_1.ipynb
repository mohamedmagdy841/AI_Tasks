```python
class Library:
    
    def __init__(self,listofbooks,library_name):
        self.list=listofbooks
        self.name=library_name
           
    def Displaybook(self):
        print("The number of available books is :",len(self.list))
        for i in self.list:
            print(i)
        
    def Lendbook(self,lend):
        if lend not in self.list:
            print("The book is not available now")
        else:
            self.list.remove(lend)
            print("You have borrowed this book:",lend)
        
    def Addbook(self,add):
        self.list.append(add)
        print("Book has been added successfully")
        
    def Returnbook(self,Return):
        self.list.append(Return)
        print("Book returned,Thank you")
        
if __name__ == "__main__":
    
    HarryLibrary=Library(["Math","Science","History"], "Harry Library")
    print("Welcome to Harry Library\n")
    
    while(True):
        print("1-Display books")
        print("2-Borrow a book ")
        print("3-Add a book")
        print("4-Return a book")
        print("5-Exit Library")
        
        user_response=int(input("Enter your choice:"))
        print("-----------------------")

        if user_response==1:
            HarryLibrary.Displaybook()
            print("-----------------------")
        elif user_response==2:
            HarryLibrary.Lendbook(input("Name of the book:"))
            print("-----------------------")
        elif user_response==3:
            HarryLibrary.Addbook(input("Name of the book:"))
            print("-----------------------")
        elif user_response==4:
            HarryLibrary.Returnbook(input("Name of the book:"))
            print("-----------------------")
        elif user_response==5:
            print("Thank You")
            break
        else:
            print("Invalid input")
            print("-----------------------")
            
        
```
